```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df16=pd.read_csv('Cases against Police Personnels/2016 Cases against Police Personnels.csv')
df17=pd.read_csv('Cases against Police Personnels/2017 Cases against Police Personnels.csv')
df18=pd.read_csv('Cases against Police Personnels/2018 Cases against Police Personnels.csv')
```

```python
df16.drop(columns='S. No.',inplace=True)
```

```python
df16.head(5)
```

```python
filt=df16['Total Number of Criminal Cases Registered during the year'].nlargest(10).index
gr16=df16.iloc[filt]
filt=df17['Number of Cases - Registered'].nlargest(10).index
gr17=df17.iloc[filt]
filt=df18['Number of Cases - Registered'].nlargest(10).index
gr18=df18.iloc[filt]
```

```python
width=0.25
x_indexes=np.arange(len(gr16['State/UT']))
plt.bar(x_indexes-width,gr16['Total Number of Criminal Cases Registered during the year'],label='2016',width=width)
plt.bar(x_indexes,gr17['Number of Cases - Registered'],label='2017',width=width)
plt.bar(x_indexes+width,gr18['Number of Cases - Registered'],label='2018',width=width)
plt.xticks(ticks=x_indexes,labels=gr16['State/UT'])
plt.xlabel('States')
plt.ylabel('No. of cases Registered')
plt.title('Cases against Police Personnels(Top 10 states)')
plt.legend()
plt.xticks(rotation=90)
```

```python
df16.head(4)
```

```python
filt2=df16.iloc[(df16['Police Personnel Arrested'].nlargest(10).index)]
plt.plot(filt2['State/UT'],filt2['Police Personnel Arrested'],label='Arrested')
plt.plot(filt2['State/UT'],filt2['Cases Chargesheeted'],label='Chargesheeted')
plt.xticks(rotation=60)
plt.legend()
plt.title('2016 top 10 states Arrested v/s Chargesheeted')
```

```python
filt2=df17.iloc[(df17['Number of Police Personel - Arrested'].nlargest(10).index)]
plt.plot(filt2['State/UT'],filt2['Number of Police Personel - Arrested'],label='Arrested')
plt.plot(filt2['State/UT'],filt2['Number of Police Personel - Chargesheeted'],label='Chargesheeted')
plt.xticks(rotation=60)
plt.legend()
plt.title('2017 top 10 states Arrested v/s Chargesheeted')
```

```python
filt2=df18.iloc[(df18['Number of Police Personel - Arrested'].nlargest(10).index)]
plt.plot(filt2['State/UT'],filt2['Number of Police Personel - Arrested'],label='Arrested')
plt.plot(filt2['State/UT'],filt2['Number of Police Personel - Chargesheeted'],label='Chargesheeted')
plt.xticks(rotation=60)
plt.legend()
plt.title('2018 top 10 states Arrested v/s Chargesheeted')
```
